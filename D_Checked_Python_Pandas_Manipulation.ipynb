```python
# Create DataFrame with None/Null to work with examples
import pandas as pd
import numpy as np
technologies   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas",None,"Spark","Python"],
    'Fee' :[22000,25000,23000,24000,np.nan,25000,25000,22000],
    'Duration':['30day','50days','55days','40days','60days','35day','','50days'],
    'Discount':[1000,2300,1000,1200,2500,1300,1400,1600]
          })
row_labels=['r0','r1','r2','r3','r4','r5','r6','r7']
df = pd.DataFrame(technologies, index=row_labels)
print(df)
```

```python
df.info()
```

# **1. Modify Column Values**

### **1.1 ใช้ .apply() เพื่อเปลี่ยนแปลงค่าของ Column ทั้งที่มีการวางเงื่อนไขในการเปลี่ยน และเปลี่ยนทุกค่าใน Columns**

```python
df['Fee'] = df['Fee'].apply(lambda x: (50/100)*x) #เปลี่ยนแปลงทุกค่าใน Column 'Fee'
```

```python
df
```

```python
#df[df['Duration'] == '']
```

```python
#มีการวางเงื่อนไขในการเปลี่ยนแปลง ดังนั้นจึงเกิดการเปลี่ยนแปลงบางค่าใน Column 'Duration'

df['Duration'] = df['Duration'].apply(lambda x: None if x == '' else x )
```

```python
df
```

```python
df.info()
```

### **1.2 ใช้ .apply() เพื่อเปลี่ยนแปลงค่าของ Column อย่างมีเงื่อนไขในการเปลี่ยนแปลง โดยเงื่อนไขนั้นเกี่ยวข้องกับการใช้ Regular Expression**

```python
import re
```

```python
# ประกาศฟังก์ชั่น เพื่อตรวจสอบว่า Argument 'origin' ที่ได้รับมานั้น มีค่าเป็น Digi หรือไม่
# ถ้ามีให้ดึงค่าเหล่านั้นออกมาทั้งหมด แล้ว return Digi กลับไป
# ถ้าไม่มี ให้ return Argument กลับไป

def extract_digi(origi):
  return re.search('\d+',origi)[0] if origi else origi 
```

```python
df['Duration'] = df['Duration'].apply(lambda x: extract_digi(x))
```

```python
df
```

```python
df.info()
```

# **6. Sort Columns & Rows**

```python
df
```

```python
df.sort_index(axis=1, ascending=False) #เรียง Columns จากมากไปหาน้อย
```

```python
# เรียง Row โดยใช้ Columns 'Courses' เป็นเงื่อนไขในการเรียง จากมากไปหาน้อย
# และเปลี่ยน Row Index Label ให้เป็น Default Label

df.sort_values(by='Courses', ignore_index=True, ascending=False)
```

```python
# เรียง Row โดยใช้ Columns 'Courses' และ 'Courses_Fee' 
# เป็นเงื่อนไขในการเรียง จากมากไปหาน้อย ตามลำดับ
# และเปลี่ยน Row Index Label ให้เป็น Default Label

df.sort_values(by=['Courses','Courses_Fee']\
               , ignore_index=True, ascending=[False,False])
```

```python
# ให้เหมือนกับก่อนหน้านี้ ยกเว้น มีการนำ NONE ซึ่งก็คือ Null ไว้ที่ Row แรกเสมอ

df.sort_values(by=['Courses','Courses_Fee']\
               , ignore_index=True, ascending=[False,False], na_position= 'first')
```

```python
# ให้เหมือนกับก่อนหน้านี้ ยกเว้น สลับเงื่อนไขให้การเรียงเป็น 'Courses_Fee' และ 'Courses'

df.sort_values(by=['Courses_Fee','Courses']\
               , ignore_index=True, ascending=[False,False], na_position= 'first')
```
